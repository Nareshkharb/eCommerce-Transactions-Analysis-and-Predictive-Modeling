```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#adding whitegrid to plot
sns.set(style = "whitegrid")
```

```python
# loading dataset
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")
```

```python
print("Customers Dataset:\n", customers.head())
print("Products Dataset:\n", products.head())
print("Transactions Dataset:\n", transactions.head())
```

```python
print("\n Missing Values:")
print("Customers:", customers.isnull().sum())
print("Products:", products.isnull().sum())
print("transactions:", transactions.isnull().sum())
```

```python
#creating dataset by merging data
merged_data = transactions.merge(customers, on = "CustomerID").merge(products, on = "ProductID")
```

```python
merged_data['TransactionDate'] = pd.to_datetime(merged_data['TransactionDate'])
merged_data['SignupDate'] = pd.to_datetime(merged_data['SignupDate'])
```

```python
merged_data
```

```python
print("\nDescriptive Statistics:\n", merged_data.describe())
```

```python
#EDA and Visualoization
# Customers Distribution by region
plt.figure(figsize=(8,5))
region_counts = customers['Region'].value_counts()
region_counts.plot(kind='bar', color='skyblue')
plt.title('Customers Distribution by Region')
plt.ylabel('Number of Customers')
plt.xlabel('Region')
plt.show()
```

```python
# Monthly Sales Trends
```

```python
monthly_sales = merged_data.groupby(merged_data['TransactionDate'].dt.to_period('M'))['TotalValue'].sum()
plt.figure(figsize=(10,6))
monthly_sales.plot(kind = 'line',color='orange')
plt.title('Monthly Sales Trends')
plt.ylabel('Total Sales(USD)')
plt.xlabel('Month')
plt.show()
```

```python
# Top 10 products by Sales
top_products = merged_data.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_products.plot(kind = 'bar', color = 'green')
plt.title('Top 10 Products by Sales')
plt.ylabel('Total Sales(USD')
plt.xlabel('Product Name')
plt.xticks(rotation = 55)
plt.show()
```

```python
#Sales by Product Category
category_sales = merged_data.groupby('Category')['TotalValue'].sum()
plt.figure(figsize=(7,7))
category_sales.plot(kind = 'pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Sales by Category')
plt.ylabel('')
plt.show()
```

```python
#Regional Sales Performance
regional_sales = merged_data.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,6))
regional_sales.plot(kind = 'bar', color='purple')
plt.title('Regional Sales Performance')
plt.ylabel('Total Sales(USD)')
plt.xlabel('Region')
plt.xticks(rotation = 0)
plt.show()
```

```python
#Saving Data
merged_data.to_csv("Cleaned_merged_data.csv", index=False)
print("Cleaned and Merged dataset saved as 'Cleaned_merged_data.csv'.")
```
